````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
renda = pd.read_csv('previsao_de_renda.csv')
renda.dtypes
````

````python
media_imovel = renda.groupby('posse_de_imovel')['renda'].mean()
media_veiculo = renda.groupby('posse_de_veiculo')['renda'].mean()
````

````python
media_imovel.to_frame()
````

````python
media_veiculo.to_frame()
````

````python
plt.figure(figsize=(6, 4))
sns.pointplot(data=renda, 
              x="posse_de_imovel", 
              y="renda", 
              errorbar=("ci", 95))
plt.title("Média de Renda por Posse de Imóvel")
plt.ylabel("Renda Média")
plt.xlabel("Possui Imóvel")
plt.show()
````

````python
plt.figure(figsize=(6, 4))
sns.pointplot(data=renda, 
              x="posse_de_veiculo", 
              y="renda", 
              errorbar=("ci", 95))
plt.title("Média de Renda por Posse de Veículo")
plt.ylabel("Renda Média")
plt.xlabel("Possui Veículo")
plt.show()
````

# INSIGHT

Analyzing the graphs and averages, we can infer that both vehicle ownership and property ownership are good predictors of income. And, statistically speaking, both discriminate against income as significant.
